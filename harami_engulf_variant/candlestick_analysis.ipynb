```python
import pandas as pd

df = pd.read_csv('niftynext50_strategy_returns.csv', names = ['zeros', 'Company', 'Max_funds', 'Min_funds', 'Date', 'Hit_rate', 'Returns'])
df = df.drop('zeros', axis = 1)
df
```

```python
def write_returns_txt(df, i):
  df['Max_funds'] = pd.to_numeric(df['Max_funds'], downcast = 'float')
  df['Min_funds'] = pd.to_numeric(df['Min_funds'], downcast = 'float')
  df['Hit_rate'] = pd.to_numeric(df['Hit_rate'], downcast = 'float')
  df['Returns'] = pd.to_numeric(df['Returns'], downcast = 'float', errors = 'coerce')
  print(df.head())

  date = df.iloc[0, 3]
  df1 = df.loc[df['Date'] == date]
  df1 = df1.reset_index()
  df1

  df2 = df.loc[df['Date'] != date]
  df2 = df2.set_index('Date')

  df_mean = df2.groupby('Company').mean()
  df_stdev = df2.groupby('Company').std()
  df1['Signal'] = np.array(df_mean['Returns'] > 0.5 * df_stdev['Returns'])
  #print(df1)

  df_ret = df1[(df1['Returns'] > 0) & df1['Signal']]
  df_sig = df1[df1['Signal']]

  string1 = ('Your strategy would have worked for ' + str(round(len(df_ret) / len(df_sig) * 100, 2)) + "% percent times in last years's " + str(len(df_sig)) + " companies.\n" )
  string2 = ("Average returns per company for them are " + str(round(df_ret['Returns'].mean(), 2)) + '% percent.\n')
  L = [date, '\n', string1, string2, '\n']

  summary = open("niftynext50_results.txt","a")
  summary.writelines(L)
  summary.close()

  return i
```

```python
# For index(sp500 or nifty50 or nifty next 50, etc) data over last 10 years to analyze strategy
dates = df['Date'].unique()
dates = dates[::-1]
print(dates) # check if dates are string in increasing order
for i in range(5):
  arr = dates[i : i + 6]
  df_new = df.loc[df['Date'].isin(arr)]
  print(write_returns_txt(df_new, i))
```
